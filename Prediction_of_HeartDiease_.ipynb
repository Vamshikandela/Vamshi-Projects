Importing Libraries


```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
import warnings
warnings.filterwarnings('ignore')

```

```python
#uploding csv data into pandas
heart_data=pd.read_csv('/content/heart_disease_data.csv')

```

```python
#read the csv data
# to get the first five records of the dataset
heart_data.head()
# to get the last five records of the dataset
heart_data.tail()
```

```python
# to get number of rows and columns in a dataset
#heart_data.shape
# to get info of the dataset
heart_data.info()
#to know the stastistical info
heart_data.describe()
```

0--- Healthy Hearts
1---Defective Hearts

```python
#splitting and
X=heart_data.drop(columns="target",axis=1)
Y=heart_data['target']
print(X,Y)
```

```python
#spllintin the data into train and test data
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,stratify=Y)
print(X_test.shape,X_train.shape)
print(Y_test.shape,Y_train.shape)
```

```python
#LogisticRegression
model=LogisticRegression()
```

```python
#model evaluation
#Accuracy Score on train data
X_trainprediction =model.fit(X_train,Y_train).predict(X_train)
Accuracyontraindata=accuracy_score(X_trainprediction,Y_train)
print("Accuracy of trained data:",Accuracyontraindata)
```

```python
#Accuracy on test data

X_testprediction= model.predict(X_test)
Accuracyontestdata=accuracy_score(X_testprediction,Y_test)
print("Accuracy of tested data:",Accuracyontestdata)
```

Building a predictive system

```python
#to Know whether the prediction  is workin or not
#taking the data

inputdata=(43,1,0,120,177,0,0,120,1,2.5,1,0,3)
#converting the data into numpy array
inputdataasarray=np.asarray(inputdata)
#reshaping to predict one data Record
reshapedata=inputdataasarray.reshape(1,-1)
prediction=model.predict(reshapedata)
print(prediction)
if (prediction[0==0]):
  print("The person heart is Healthy ....")
else:
  print("The person heart is Defective")
```
