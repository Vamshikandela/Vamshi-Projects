```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn import svm
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import warnings
warnings.filterwarnings('ignore')
```

```python
#collecting and Analysis of data set
Diabetes_dataset=pd.read_csv('/content/diabetes.csv')
#To get the first five records fromthe dataset
Diabetes_dataset.head()
#To get the last five records fromthe dataset
Diabetes_dataset.tail()
```

```python
#to know the No of rows are there in the Dataset
Diabetes_dataset.shape
```

```python
#getting the stastical methods of dataset
Diabetes_dataset.describe()
```

```python
#to know the null values in dataset
Diabetes_dataset.isnull().sum()
```

```python
# to know How many 1and 0 are in dataset

Diabetes_dataset['Outcome'].value_counts()
```

0--non diabetes
1--diabetes

```python
#finding mean of diabetes and non diabetes
Diabetes_dataset.groupby('Outcome').mean()
```

```python
#seperate the data into two
X=Diabetes_dataset.drop(columns='Outcome',axis=1)
Y=Diabetes_dataset['Outcome']
#knowing x and y Values
print(X)
#print(Y)
```

data Strandiadised

```python
#import the StandardScaler into scaler variable
scaler=StandardScaler()
```

```python
#Xdata fit in the scaler
scaler.fit(X)
```

```python
#transform the data X into scaler
standard_data=scaler.transform(X)
```

```python
print(standard_data)
```

```python
# convert data into X and Y
X=standard_data
Y=Diabetes_dataset['Outcome']
```

```python
#Model Evaluation
#into X_train and Y_train and 20% of test (X_test,Y_test)
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,stratify=Y,random_state=2)
```

```python
X_train.shape
Y_train.shape
```

Training the Model

```python
from sklearn import svm
classifier=svm.SVC(kernel='linear')
```

```python
classifier.fit(X_train,Y_train)
```

Model Evaluation


```python
#accuracy score of trained data
from sklearn.metrics import accuracy_score
X_train_predicton=classifier.predict(X_train)
trained_accuracy=accuracy_score(X_train_predicton,Y_train)
print("trained_data_accuracy :",trained_accuracy )
```

```python
#accuracy score of tested data
from sklearn.metrics import accuracy_score
X_test_predicton=classifier.predict(X_test)
tested_accuracy=accuracy_score(X_test_predicton,Y_test)
print("tested_data_accuracy:",tested_accuracy)
```

```python
#Checking the predective system of model
inputdata=(5,116,74,0,0,25.6,0.201,30)
#changing the data into numpy array to access one record
import numpy as np
input_data_as_nparray=np.asarray(inputdata)
#reshape the data for one record
reshaped_data=input_data_as_nparray.reshape(1,-1)
#this reshaped array transform to classifier
std_data=scaler.transform(reshaped_data)
print(std_data)
#predict the Input data whether a person is diabetic or non-diabetic
prediction=classifier.predict(std_data)
print(prediction)
#output gives with the conditional whether person is diabetic or not
if (prediction[0==0]):
  print("This person is non-diabetic")
else:
  print("This person is diabetic")
```
